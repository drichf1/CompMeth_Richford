```python
# Homework Exercise 2.6
# distance of closest approach l1
#linear velocity at perihelion v1 (aphelion l2, v2)
# Kepler: l2*v2 = l1*v1
# a.  E = 1/2 m v**2 - GmM/r
#     Show v2 is the smaller root of the quadratic eqn
#
#     v2**2 - (2*G*M*v2)/(l1*v1) - (v1**2 - (2*G*M)/l1) = 0

import numpy

def quadraticformula(a,b,c):
    added_root = (-b+numpy.sqrt(b**2.-4.*a*c))/(2.*a)
    subtracted_root = (-b-numpy.sqrt(b**2.-4.*a*c))/(2.*a)
    return numpy.array([added_root,subtracted_root])

def aphelion(l1,v1,v2):
    l2 = (l1*v1)/v2
    return l2

# Use the proper Root
def checkroot(l1, v1, radius, roots): #takes array for roots
    #centripital force = gravitational force
    G = 6.67*10**(-11) #m**3/kg*s**2
    M = 1.9881*10**30 #kg
    circular_velocity = numpy.sqrt(M*G/l1) # circular orbit
    escape_velocity = numpy.sqrt(2*M*G/l1) # escape veloc.
    if (l1 <= radius):
        print("Oh, no! THe satellite will crash!")
        return 0
    if (v1 >= escape_velocity):
        print("Oh, no! The satellite will not make an orbit!")
        return 0
    if (v1 == circular_velocity):
        print("Oh-ho! Orbit is circular!")
        return 0
    if (v1 < circular_velocity):
        print("perihelion!")
        v2 = roots[1]
        return v2
    elif (v1 > circular_velocity):
        print("aphelion!")
        v2 = roots[0]
        return v2

# Calculate Roots
v1= 1.*10**3
l1= 1.*10**6
r = 1.*10**3
G = 6.67*10**(-11) #m**3/kg*s**2
M = 1.9881*10**30 #kg

a = 1.
b = -(2.*G*M)/(l1*v1)
c = -(v1**2. - (2.*G*M)/l1)

roots = quadraticformula(a,b,c)
print("roots")
print(roots)

# Perihelion or Aphelion
v2 = 0.
v2 = checkroot(l1, v1, r, roots)

# calculate Aphelion
if v2 != 0:
    l2 = aphelion(l1,v1,v2)
    print("l2")
    print(l2)
else:
    print("error")
```
